```python
import pandas as pd
import numpy as np
import warnings
```

```python
pd.set_option('display.max_rows', 500)
```

```python
def generate_fold_expression(dataset, reference_columns, timepoints):
    avg_conditions = controls_brief_in_vivo[reference_columns]
    for condition in conditions:
        with warnings.catch_warnings():
            warnings.simplefilter('ignore')
            avg_conditions[condition] = dataset.filter(like=condition).mean(axis=1)
            
    fold_dataset = avg_conditions[reference_columns]
    for timepoint in timepoints:
        temp = avg_conditions.filter(like=timepoint)
        with warnings.catch_warnings():
            warnings.simplefilter('ignore')
            fold_dataset[timepoint] = np.array(temp.filter(like='light'))/np.array(temp.filter(like='dark'))
        
    return fold_dataset, avg_conditions
    
```

```python
brief_in_vivo = pd.read_csv("/Users/kbw29/Dropbox/Research/Neurobiology_PhD/Huang/Projects/CellReadR/Data/Reference/Tyssowski-2018_in-vivo-brief.csv")
brief_in_vivo
```

```python
sustained_in_vivo = pd.read_csv("/Users/kbw29/Dropbox/Research/Neurobiology_PhD/Huang/Projects/CellReadR/Data/Reference/Tyssowski-2018_in-vivo-sustained.csv")
sustained_in_vivo
```

```python
reference_columns = ['gene name', 'Gene Class']
conditions = ['30m dark', '30m light', '150m dark', '150m light']
timepoints = ['30m', '150m']
controls_brief_in_vivo = brief_in_vivo[reference_columns + list(brief_in_vivo.filter(like='corn oil').columns)]

fold_brief_in_vivo, avg_brief_in_vivo = generate_fold_expression(controls_brief_in_vivo, reference_columns, timepoints)
fold_brief_in_vivo
```

```python
conditions = ['dark 30m', '30m light', '150m dark', '150m light']
controls_sustained_in_vivo = sustained_in_vivo[reference_columns + list(sustained_in_vivo.filter(like='corn oil').columns)]

fold_sustained_in_vivo, avg_sustained_in_vivo = generate_fold_expression(controls_sustained_in_vivo, reference_columns, timepoints)
fold_sustained_in_vivo
```
